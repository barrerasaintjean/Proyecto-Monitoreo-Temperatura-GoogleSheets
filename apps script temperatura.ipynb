Al revisar los primeros datos, noté algo clave: mi sensor marcaba entre 3°C y 6°C más que el panel de la cámara. La razón es física: el sensor de la máquina mide donde sale el aire frío, pero el mío está cerca de la puerta (el punto más crítico). Este hallazgo me enseñó que no basta con medir, hay que saber dónde medir para entender la temperatura real que está sufriendo el producto al abrir y cerrar la puerta.

```python
function doGet(e) {
  var ss = SpreadsheetApp.getActiveSpreadsheet();
  var sheet = ss.getSheets()[0];
  
  // Capturamos los datos con los nombres exactos que usa el ESP32
  var temp1 = e.parameter.t_conge;
  var temp2 = e.parameter.t_amb;
  var fecha = new Date();

  // Si los datos llegan vacíos, escribimos un error para saber qué pasó
  if (!temp1 || !temp2) {
    sheet.appendRow([fecha, "Error", "Sin Datos"]);
    return ContentService.createTextOutput("Error: No llegaron parámetros");
  }

  // Insertamos los datos asegurándonos de que Excel los vea como números
  sheet.appendRow([fecha, parseFloat(temp1), parseFloat(temp2)]);
  
  return ContentService.createTextOutput("Éxito: Datos guardados");
}
```
